```python
import pandas as pd
```

```python
dataset_in_service = pd.read_csv('https://docs.google.com/spreadsheets/d/' + 
                   '1DR72UewokRt_04MOFr0Ti3FrynkYIjct' +
                   '/export?gid=230400863&format=csv', keep_default_na=False)
dataset_in_service
```

```python
datasets = pd.read_csv('https://docs.google.com/spreadsheets/d/' + 
                   '1jp5us3v7YRfPdCAUXfO8DXFyfsTjv5wr' +
                   '/export?gid=797750865&format=csv', keep_default_na=False)
datasets
```

```python
services = pd.read_csv('https://docs.google.com/spreadsheets/d/' + 
                   '17JjkDXnoI3-BvgBo8Ku6e8HBCd8MXhZZ' +
                   '/export?gid=1391930590&format=csv', keep_default_na=False)
services
```

```python
full_datasets_in_service = pd.merge(dataset_in_service, datasets, left_on=["dataset_fk"], right_on=["dataset_internal_id"], how="left", sort=False)
full_datasets_in_service
```

```python
full_datasets_in_services = pd.merge(full_datasets_in_service, services, left_on=["service_fk"], right_on=["service_ID"], how="left", sort=False)
full_datasets_in_services
```

```python
#remove leading/tailing whitespaces in all columns
full_datasets_in_services = full_datasets_in_services.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
columns = ['ALG','ANI','BAC','CYA','FUN','PLA','VIR','EUR','AT', 'BE', 'BG', 'HR', 'CZ', 'CY', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU', 'IS', 'IE', 'IT', 'LV', 'LT', 'LU', 'MT', 'NL', 'NO', 'PL', 'PT', 'RO', 'SK', 'SI', 'ES', 'SE', 'CH', 'GB']
full_datasets_in_services[columns] = full_datasets_in_services[columns].replace(-1, True).values
full_datasets_in_services[columns] = full_datasets_in_services[columns].replace(0, False).values
full_datasets_in_services[columns]
```

```python
uri_columns = ['dataset_description_uri', 'download_website_uri', 'homepage_uri', 'query_uri', 'access_conditions_uri', 'service_description_uri', 'file_matching_uri', 'api_documentation_uri', 'api_endpoint_uri', 'open_refine_reconciliation_if_uri', 'open_refine_documentation_uri']
full_datasets_in_services[uri_columns] = full_datasets_in_services[uri_columns].replace(r"^;", r"", regex=True)
full_datasets_in_services[uri_columns] = full_datasets_in_services[uri_columns].replace(r"^--$", r"", regex=True)
full_datasets_in_services[uri_columns]
```

```python
def column_to_tag_group(record, column_names, tag_names=None):
    tags = []
    for index, column_name in enumerate(column_names):
        if(record[column_name]):
            if(tag_names is None):
                tags.append(column_name)
            else:
                tags.append(tag_names[index])
    return tags
    

records = []
for row_index, row in full_datasets_in_services.iterrows():
    organism_groups = column_to_tag_group(row, ['ALG','ANI','BAC','CYA','FUN','PLA','VIR','marine_only','fossil_only','contains_fossils','biota'],['algae','animalia','bacteria','cyanobacteria','fungi','plantae','virus','marine_only','fossil_only','contains_fossils','multiple'])
    geographic_scope = column_to_tag_group(row, ['global','EUR','AT', 'BE', 'BG', 'HR', 'CZ', 'CY', 'DK', 'EE', 'FI', 'FR', 'DE', 'GR', 'HU', 'IS', 'IE', 'IT', 'LV', 'LT', 'LU', 'MT', 'NL', 'NO', 'PL', 'PT', 'RO', 'SK', 'SI', 'ES', 'SE', 'CH', 'GB'],['global', "europe",'at', 'be', 'bg', 'hr', 'cz', 'cy', 'dk', 'ee', 'fi', 'fr', 'de', 'gr', 'hu', 'is', 'ie', 'it', 'lv', 'lt', 'lu', 'mt', 'nl', 'no', 'pl', 'pt', 'ro', 'sk', 'si', 'es', 'se', 'ch', 'gb'])
    purpose = column_to_tag_group(row, ['full_name_checked','output_taxonomic_status','output_classification'],['full_name_checked','acceptance','classification'])
    linkage = column_to_tag_group(row, ['output_persistent_id','output_persistent_resolvable_id','name_id_resolved_2_taxon'])
    requirements = column_to_tag_group(row, ['direct_matching_interface','file_matching_interface','open_refine_reconciliation_if','download_full_from_website','download_dataset','api_public_access','R-package','local_installation'], ['direct_matching_interface','file_matching_interface','open_refine_reconciliation_if','download_full_from_website','download_dataset','api','r_package','local_installation'])
    direct_matching_restriction = int(row.direct_matching_restriction) if str.isdigit(row.direct_matching_restriction) else 0
    file_matching_restriction = row.file_matching_restriction
    record = {'id':row_index,'service_name':row.service_name_y, 'dataset_name':row.display_title, 'dataset_short_name':row.short_title_y, 'notes':row.notes, 'caveats':row.caveats, 'organism_groups':organism_groups, 'geographic_scope':geographic_scope, 'purpose':purpose, 'linkage':linkage, 'requirements':requirements, 'homepage':row.homepage_uri, 'api_documentation':row.api_documentation, 'direct_matching_restriction':direct_matching_restriction, 'file_matching_restriction':file_matching_restriction}
    records.append(record)
    
records[0]
```

```python
import json

json_output = json.dumps(records, indent=4)

with open('../data.js', 'w') as f:
    f.write('var rows = ')
    f.write(json_output)
```
